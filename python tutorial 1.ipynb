**mutable means we can change the value of it**

**Lists are mutable**

```python
nums.append(45)       #will append(add) 45 to the list nums at the end
nums.insert(1,88)     #insert function is used to insert values to a list at a particular index 
#unlike append which inserts the value at the end in insert function the 1st input is the index where we are  adding the element and 
#2nd input is the value that is being inserted 
print(nums)
nums.remove(45)       #it removes the value 45 from the list
nums.pop(2)           #this function is used to pop(remove) elements of the list based on the index print(nums)//number passed in to the function.
print(nums)         #every time you run this code 45 & 88 will be added to the list at the end an at the index 1 respectively
```

```python
num = [11,2,1,57,8] #created another list num
num.pop()        #if the index of the element to be poped is not specified it pops the last element of the list
print(num)
del num[2:]      #delete elements of the list from index 2 till the end
print(num)
num.extend([7, 88, 8, 56])  # will add multiple elements to the list at the end
print(num)
print(min(num))  #min number in the list
print(max(num))  #max number in the list
print(sum(num))  #sum of the number in the list
num.sort()       #this function will sort all the elements of the list
print(num)
```

**Tuple & Set**

Tuple is almost same as list but **tuple is immutable**

```python
#tuple
tup = (21,36,14,25,36)  #we use ()for tuples and[] for lists
print(tup[2])        #indexing is same
#tup[1]=33  #this will be shown as error because tuple is immutable
print(tup.count(36)) #prints the no.of times the input element is occuring in the tuple
print(tup.index(21)) #gives the index of the input element
print(tup.index(21,1,3)) #searches for the element 21 from index 1 to 3 of the tuple and if the element 21 is not in index from 1 to 3 then it will return error x not in tuple
#iteration in tuples is faster than in lists as we will not add any new elements
#we use tuple to enhance the speed of our execution
```

*Set is an unordered collection of data type The elements in the set cannot be duplicates. The elements in the set are immutable(cannot be modified) but the set as a whole is mutable.*

The major advantage of using a set, as opposed to a list, is that it has a highly optimized method for checking whether a specific element is contained in the set.



```python
#set
s = {45,87,88,21,10,22,25,14,21,5} #sets are un-ordered
print(s) #set will use hash to fetch the element in set as fast as possible thats why it just goes with the flow instead of oredering its elements
#it will not support duplicate values
```

```python
set1 = {'s','t'}
print('x' in set1)
print('s' in set1)
set2 = {'s'}
print('s' in set2)
```

Lists cannot be added to a set as elements because Lists are not hashable whereas Tuples can be added because tuples are immutable and hence Hashable.

```python
data ={1:'sai',2:'sunny',5:'bobby'} #instead of index we give specific key to each element syntax{key:'value'}
print(data[5])    #fetching data from dictionaries keys of the dictionary need not be integers it can also be a string
print(data.get(4))
print(data.get(4,'not found'))  #here the message passed btw'' will be displayed in case the given key doen't exsist
#creating dict with lists
keys=['sai','teja','palachuri'] #creating two lists
vals=['python','c','c++']

data = dict(zip(keys,vals))   #combining two lists into a dict using zip command the first list passed will be stored as keys and 2nd list passed will be stored as elements of the dictionary

print(data)
print(data['sai']) #calling for the values of the dict by mentioning keys eg. sai is key and python is the value
data['sunny']='java' #adding elements to the dict 
#syntax to add elements to the dictionary
# nameOfTheDict['key']='value'
print(data)
del data['palachuri']
# syntax to delete a element from the dict is
# del NameoftheDict['key']
print(data)
#we can build a list as well as a dictionary inside a dictionary
prog={'mom':'swathi','bobby':['sumanth','saiteja'],'sunny':{'firstname':'mohan vamsi','lastname':'palachuri'}}
print(prog['sunny']) # printing the value for key 'sunny' in prog dictionary here the output is also a string
print(prog['sunny']['lastname'])  #printing the value for key 'sunny' in prog dictionary with key 'lastname' as key in sunny dictionary
print(prog['bobby'][1])       #printing the value for key 'bobby' in prog dictionary and list index 1
```

WHILE LOOPS

```python
i=1
while i<=5:    #iterating while incrementing i
  print('Sai Teja',i)
  i=i+1   # can also be written as i+=1
```

```python
i=3
while i>=1:   #iterating while derementing i
  print('python ',end='')  # in python when we use print function it automatically prints the statement in the new line but here we want to print satatement of the inner while loop in the same line as the print statement of the outer while loop
  #when we use end='' or end="" function in the print then after printing the outpout stays in the same line
  j=1  #we have to declare the variable of inner loop inside the first loop itself so that its value will be set back to 1 after completing inner while loop
  while j<=3:
    print('rocks ',end='') # using end='' function so that we get output like rocks rocks rocks in the same line
    j=j+1
  i=i-1
  print()  # as we have used a end="" function in the previous print we use a print() function so that when while loop runs again for the i=i-1 we print the output in a new line
  #the above print() function is just = printf("\n") in C programming
  #here we need not use \n becoz python automatically goes to next line for every print() function but if we use \n again in the print() then we get an empty line between the two print statements in the loops.
```
